Descrevendo as Características(Features):

1 --> Sim
0 -->Não

- Ter Pelo Longo
- Ter Perna Curta
- Fazer Auau

Essa será nossa Tabela com os Dados:

```python
porco1 = [0,1,0]
porco2 = [0,1,1]
porco3 = [1,1,0]

cachorro1 = [0,1,1]
cachorro2 = [1,0,1]
cachorro3 = [1,1,1]

treino_x = [porco1,porco2,porco3,cachorro1,cachorro2,cachorro3]

#1: Porco , 0: Cachorro
treino_y = [1,1,1,0,0,0] #Classes ou Labels
```

Vamos importar a Biblioteca de treinamento e treinar esses dados:

```python
from sklearn.svm import LinearSVC

model = LinearSVC()
model.fit(treino_x, treino_y)
```

Vamos colocar um animal misterioso e ver a previsão do algoritmo:

```python
animal_misterioso =[1,1,1]
model.predict([animal_misterioso])
```

Agora vamos fazer uma série de Testes:

```python
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]

teste_x = [misterio1,misterio2,misterio3]
previsoes = model.predict(teste_x)
```

As respostas dos testes são:

```python
teste_y = [0,1,1]
```

A Taxa de Acerto do Algoritmo é:

```python
corretos = (previsoes == teste_y).sum()
total = len(teste_y)
acertos = corretos/total

print("Taxa de acerto:", acertos*100)
```

A Taxa de Acerto do Algoritmo é:

```python
from sklearn.metrics import accuracy_score

taxa_de_acerto =accuracy_score(teste_y,previsoes)
print("Taxa de acerto:",acertos*100)
```
